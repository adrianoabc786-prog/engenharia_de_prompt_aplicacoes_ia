```python
def verificar_temperatura_sistema(temperatura):
    """
    Verifica a temperatura do sistema e exibe um alerta de superaquecimento
    se a temperatura for maior que 30 graus Celsius.

    Args:
        temperatura (int): A temperatura atual do sistema em graus Celsius.
    """
    if temperatura > 30:
        # Exibe um alerta se a temperatura exceder 30°C
        print("ALERTA: Superaquecimento! Temperatura atual: {}°C".format(temperatura))
    else:
        # Informa que a temperatura está normal
        print("Temperatura do sistema está normal: {}°C".format(temperatura))

# --- Exemplo de uso com entrada do usuário ---
try:
    # Solicita ao usuário para inserir a temperatura
    temperatura_digitada = int(input("Por favor, digite a temperatura atual do sistema em graus Celsius: "))
    verificar_temperatura_sistema(temperatura_digitada)
except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro para a temperatura.")
```

ATIVIDADE 1 - AUTOMAÇÃO DE ARQUIVOS


```python
import os
import shutil
import random

# --- Configurações ---

# Diretório onde os arquivos bagunçados serão criados e organizados
DIRETORIO_BASE = 'arquivos_para_organizar'
DIRETORIO_BAGUNCADO = os.path.join(DIRETORIO_BASE, 'baguncados')
DIRETORIO_ORGANIZADO = os.path.join(DIRETORIO_BASE, 'organizados')

# Extensões e seus respectivos diretórios de destino
EXTENSOES_MAPA = {
    '.pdf': 'Documentos/PDFs',
    '.doc': 'Documentos/Word',
    '.docx': 'Documentos/Word',
    '.xls': 'Documentos/Excel',
    '.xlsx': 'Documentos/Excel',
    '.ppt': 'Documentos/PowerPoint',
    '.pptx': 'Documentos/PowerPoint',
    '.jpg': 'Imagens',
    '.jpeg': 'Imagens',
    '.png': 'Imagens',
    '.gif': 'Imagens',
    '.mp3': 'Musicas',
    '.wav': 'Musicas',
    '.mp4': 'Videos',
    '.avi': 'Videos',
    '.txt': 'Textos',
    '.zip': 'Compactados',
    '.rar': 'Compactados',
    '.py': 'ScriptsPython',
    '.js': 'ScriptsWeb'
}

# --- Funções ---

def criar_arquivos_exemplo(num_arquivos=12):
    """
    Cria arquivos de exemplo com extensões variadas no diretório bagunçado.
    """
    print(f"\nCriando {num_arquivos} arquivos de exemplo em '{DIRETORIO_BAGUNCADO}'...")
    os.makedirs(DIRETORIO_BAGUNCADO, exist_ok=True)

    extensoes = list(EXTENSOES_MAPA.keys()) + ['.log', '.json', '.dat'] # Adiciona algumas extras para 'Outros'

    for i in range(num_arquivos):
        nome_base = f"arquivo_{i + 1}_{random.randint(100, 999)}"
        extensao = random.choice(extensoes)
        nome_arquivo = f"{nome_base}{extensao}"
        caminho_arquivo = os.path.join(DIRETORIO_BAGUNCADO, nome_arquivo)

        # Cria um arquivo vazio
        with open(caminho_arquivo, 'w') as f:
            f.write(f"Conteúdo de {nome_arquivo}")
        print(f"  Criado: {nome_arquivo}")
    print("Arquivos de exemplo criados com sucesso!")

def organizar_arquivos():
    """
    Organiza os arquivos do diretório bagunçado em pastas baseadas na extensão.
    """
    print(f"\nOrganizando arquivos de '{DIRETORIO_BAGUNCADO}' para '{DIRETORIO_ORGANIZADO}'...")
    os.makedirs(DIRETORIO_ORGANIZADO, exist_ok=True)

    for nome_arquivo in os.listdir(DIRETORIO_BAGUNCADO):
        caminho_origem = os.path.join(DIRETORIO_BAGUNCADO, nome_arquivo)

        # Ignora diretórios, se houver algum no meio
        if os.path.isdir(caminho_origem):
            continue

        # Obtém a extensão do arquivo
        _, extensao = os.path.splitext(nome_arquivo)
        extensao = extensao.lower() # Converte para minúsculas para padronização

        # Determina o subdiretório de destino
        subdiretorio_destino = EXTENSOES_MAPA.get(extensao, 'Outros')
        caminho_destino_diretorio = os.path.join(DIRETORIO_ORGANIZADO, subdiretorio_destino)

        # Cria o diretório de destino se não existir
        os.makedirs(caminho_destino_diretorio, exist_ok=True)

        # Move o arquivo
        caminho_destino_arquivo = os.path.join(caminho_destino_diretorio, nome_arquivo)
        shutil.move(caminho_origem, caminho_destino_arquivo)
        print(f"  Movido: '{nome_arquivo}' para '{subdiretorio_destino}'")
    print("Organização de arquivos concluída com sucesso!")

def limpar_diretorios():
    """
    Remove os diretórios criados para os exemplos e organização.
    """
    print(f"\nLimpando diretórios '{DIRETORIO_BASE}'...")
    if os.path.exists(DIRETORIO_BASE):
        shutil.rmtree(DIRETORIO_BASE)
        print("Diretórios removidos com sucesso.")
    else:
        print("Diretórios de exemplo não encontrados para limpeza.")

# --- Execução do programa ---
if __name__ == "__main__":
    # Limpa diretórios de execuções anteriores para começar 'do zero'
    limpar_diretorios()

    # 1. Cria os arquivos de exemplo
    criar_arquivos_exemplo(num_arquivos=12)

    # 2. Organiza os arquivos
    organizar_arquivos()

    print("\n--- FIM DO PROGRAMA ---")
    print(f"Verifique a estrutura em: {os.path.abspath(DIRETORIO_ORGANIZADO)}")
    print("Para limpar os diretórios criados, execute a função 'limpar_diretorios()' novamente.")
```

ATIVIDADE 2 - CONSULTA SIMPLES E APIs

```python
import requests

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço a partir de um CEP.

    Args:
        cep (str): O CEP a ser consultado (apenas números).

    Returns:
        dict or None: Um dicionário com os dados do endereço se a consulta for bem-sucedida,
                      ou None se o CEP for inválido ou ocorrer um erro.
    """
    # Remove qualquer caractere não numérico do CEP
    cep = ''.join(filter(str.isdigit, cep))

    if len(cep) != 8:
        print("Erro: O CEP deve conter 8 dígitos.")
        return None

    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Levanta uma exceção para códigos de status HTTP de erro
        data = response.json()

        if "erro" in data and data["erro"]:
            print(f"Erro: CEP '{cep}' não encontrado ou inválido.")
            return None
        return data
    except requests.exceptions.RequestException as e:
        print(f"Erro de conexão ao consultar o CEP: {e}")
        return None
    except ValueError as e:
        print(f"Erro ao processar a resposta da API: {e}")
        return None

def formatar_endereco(dados_endereco):
    """
    Formata os dados de endereço retornados pela API ViaCEP para exibição legível.

    Args:
        dados_endereco (dict): Dicionário com os dados do endereço.

    Returns:
        str: Uma string formatada com as informações do endereço.
    """
    if not dados_endereco:
        return "Nenhuma informação de endereço disponível."

    endereco_formatado = [
        f"CEP: {dados_endereco.get('cep', 'N/A')}",
        f"Logradouro: {dados_endereco.get('logradouro', 'N/A')}",
        f"Complemento: {dados_endereco.get('complemento', '')}",
        f"Bairro: {dados_endereco.get('bairro', 'N/A')}",
        f"Localidade: {dados_endereco.get('localidade', 'N/A')} - {dados_endereco.get('uf', 'N/A')}",
        f"IBGE: {dados_endereco.get('ibge', 'N/A')}",
        f"GIA: {dados_endereco.get('gia', 'N/A')}",
        f"DDD: {dados_endereco.get('ddd', 'N/A')}",
        f"SIAFI: {dados_endereco.get('siafi', 'N/A')}"
    ]
    # Filtra linhas vazias para complemento
    return "\n".join([linha for linha in endereco_formatado if not linha.endswith(': ')])

# --- Exemplo de uso ---
if __name__ == "__main__":
    print("\n--- Consulta de Endereço por CEP ---")
    cep_digitado = input("Digite o CEP (apenas números, ex: 01001000): ")

    dados = consultar_cep(cep_digitado)

    if dados:
        print("\n--- Endereço Encontrado ---")
        print(formatar_endereco(dados))
    else:
        print("\nNão foi possível obter o endereço para o CEP informado.")
```

ATIVIDADE 3 - SISTEMA DE NOTIFICAÇÃO

```python
# Script Python para simular um sistema de monitoramento e disparo de alertas.
# Ele verifica uma condição crítica e notifica quando ela é atingida.

import time
import random

# --- Configurações ---

LIMITE_CRITICO = 90  # Exemplo: Uso de CPU acima de 90% é crítico
INTERVALO_VERIFICACAO = 5 # Verificar a cada 5 segundos (para simulação)

# --- Funções ---

def obter_valor_sistema_simulado():
    """
    Simula a obtenção de um valor do sistema (ex: uso de CPU).
    Em um cenário real, isso leria um sensor ou API.
    """
    return random.randint(0, 100) # Retorna um valor aleatório entre 0 e 100

def verificar_condicao_critica(valor):
    """
    Verifica se o valor obtido atingiu a condição crítica.
    """
    return valor > LIMITE_CRITICO

def disparar_alerta(mensagem):
    """
    Dispara um alerta, exibindo uma mensagem no console e simulando o envio de um e-mail.
    """
    print("\n!!! ALERTA CRÍTICO !!!")
    print("Mensagem: {}".format(mensagem))
    print("[SIMULADO] E-mail de alerta enviado para administradores.")
    print("----------------------")

# --- Programa Principal ---

if __name__ == "__main__":
    print("\n--- INICIANDO SISTEMA DE MONITORAMENTO --- ")
    print("Monitorando um valor de sistema (simulado) com limite crítico de {}%".format(LIMITE_CRITICO))

    # Loop para simular monitoramento contínuo
    # Para parar, interrompa a execução da célula.
    try:
        for i in range(5): # Simula 5 ciclos de verificação
            print("\nVerificação {} de 5...".format(i+1))
            valor_atual = obter_valor_sistema_simulado()
            print("Valor atual do sistema (simulado): {}%".format(valor_atual))

            if verificar_condicao_critica(valor_atual):
                mensagem_alerta = "O valor do sistema atingiu {}% que excede o limite crítico de {}%!".format(valor_atual, LIMITE_CRITICO)
                disparar_alerta(mensagem_alerta)
            else:
                print("Condição do sistema normal.")

            if i < 4: # Não espera após a última verificação
                print("Aguardando {} segundos para a próxima verificação...".format(INTERVALO_VERIFICACAO))
                time.sleep(INTERVALO_VERIFICACAO)

    except KeyboardInterrupt:
        print("\nMonitoramento interrompido pelo usuário.")
    finally:
        print("\n--- SISTEMA DE MONITORAMENTO FINALIZADO --- ")
```
