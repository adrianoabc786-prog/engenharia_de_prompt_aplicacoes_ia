Eh primo

```python
import math
def eh_primo(n):
  """Verifica se um número n é primo."""
  # Números menores ou iguais a 1 não são primos
  if n <= 1:
    return False

  # Percorre de 2 até n-1 para verificar divisores
  for i in range(2, int(math.sqrt(n))):
    # Se encontrar um divisor, o número não é primo
    if n % i == 0:
      return False

  # Se o loop terminar sem encontrar divisores, o número é primo
  return True
```

Inversão de string

```python
def inverte_texto(texto):
  # Esta função recebe uma string e a retorna invertida
  # Utilizamos o fatiamento (slicing) [::-1] que é a forma mais eficiente em Python
  return texto[::-1]

# Exemplo de uso:
nome = "adriano"
resultado = inverte_texto(nome)

# Exibe o resultado da inversão
print(f"Original: {nome}")
print(f"Invertido: {resultado}")
```
